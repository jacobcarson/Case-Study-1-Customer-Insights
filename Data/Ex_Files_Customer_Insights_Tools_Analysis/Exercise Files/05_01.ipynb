```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pylab as plt
%matplotlib inline
```

```python
ci_data = pd.read_csv('data/ci-data.csv')
```

```python
x = pd.DataFrame(ci_data, columns = ['purchase_sum', 'experience'])
```

```python
y = pd.DataFrame(ci_data.characteristic_one)
y.columns = ['Targets']
```

```python
colormap = np.array(['', 'red', 'black'])
plt.scatter(x.purchase_sum, x.experience, c=colormap[y.Targets])
```

```python
model = KMeans(n_clusters=2)
```

```python
model.fit(x)
```

```python
model.labels_
```

```python
colormap = np.array(['red','black'])
```

```python
plt.scatter(x.purchase_sum, x.experience, c=colormap[model.labels_])
```
